```python
import pandas as pd

censusDF = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/acs2017_census_tract_data.csv')

censusDF.County = censusDF.County.str.replace(" County", "")

newCensusDF = pd.DataFrame({'County':censusDF.groupby(['County']).State.unique()})

censusDF.head()
```

```python
assignDF = pd.DataFrame({"County":[], "State":[], "Population":[], "Poverty":[], "PerCapitaIncome":[]})

assignDF.County = censusDF.County
assignDF.State = censusDF.State
assignDF.Population = censusDF.TotalPop
assignDF.Poverty = censusDF.Poverty
assignDF.PerCapitaIncome = censusDF.IncomePerCap

assignDF.head()
```

```python

newAssignDF = assignDF.drop_duplicates(['County', 'State'])
newAssignDF.Population = assignDF.groupby(['County', 'State']).Population.transform('sum')
newAssignDF.Poverty = assignDF.groupby(['County', 'State']).Poverty.transform('mean')
newAssignDF.PerCapitaIncome = assignDF.groupby(['County', 'State']).PerCapitaIncome.transform('mean')

newAssignDF.head()
```

```python
print(newAssignDF[(newAssignDF['County'] == 'Loudoun') & (newAssignDF['State'] == 'Virginia')])
print(newAssignDF[(newAssignDF['County'] == 'Washington') & (newAssignDF['State'] == 'Oregon')])
print(newAssignDF[(newAssignDF['County'] == 'Harlan') & (newAssignDF['State'] == 'Kentucky')])
print(newAssignDF[(newAssignDF['County'] == 'Malheur') & (newAssignDF['State'] == 'Oregon')])
print(newAssignDF[newAssignDF['Population'] == newAssignDF['Population'].max()])
print(newAssignDF[newAssignDF['Population'] == newAssignDF['Population'].min()])
```
